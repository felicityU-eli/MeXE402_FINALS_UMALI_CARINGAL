```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("MeXE402_FINALS_UMALI_CARINGAL/ELex Finals/TWICE DATASET/twice1.jpg")
face_cascade = cv2.CascadeClassifier("MeXE402_FINALS_UMALI_CARINGAL/ELex Finals/Cascade Classifier/haarcascade_frontalface_default.xml") #detects face
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #convert to grayscale. why? because CascadeClassifier is design to run only grayscale images

faces = face_cascade.detectMultiScale(gray,1.3,5) #detectMultiScale(image[, scaleFactor[,minNeighbors[, flags[, minSize[, maxSize]]]]]) -> objects #detection functionn which detects image for us #play around the values if not working sa image na gamit
for (x, y, w, h) in faces:
  center_coordinates = (x + w//2, y + h//2) # calculate the center of the detected face
  axesLength = (w//2, h//2) # calculate the axes length (half of the width and height of the detected face)

  cv2.ellipse(image, center_coordinates, axesLength, 0, 0, 360, (0, 255, 0), 2) # draw an ellipse on the image

cv2_imshow(image)
```
